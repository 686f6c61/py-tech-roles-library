## Empezar

Una clase, carga perezosa. Cada idioma es un catálogo completo; los códigos (`BE-L3`) son idénticos en ambos.


```python
from tech_roles_library import TechRoles

en = TechRoles(language="en")
es = TechRoles(language="es")

role = en.get_role("BE-L3")
print(role.role, "|", role.level)
print(es.get_role("BE-L3").role)
print(role.core_competencies[0])
```

## Estadísticas del catálogo


```python
s = en.statistics()
print(s.total_roles, "roles ·", s.total_entries, "definiciones ·", s.total_categories, "categorías")
s.by_category
```

## Consultar roles

Por código o por nombre y nivel. El nivel acepta `3`, `"3"`, `"L3"` o el nombre completo.


```python
print(en.get_role("AAG-L5").role, en.get_role("AAG-L5").years_range)
print(en.get_role_by_name("Backend Developer", 3).code)
print(es.get_role_by_name("Desarrollador de Backend", "L3").code)
[e.code for e in en.get_levels("Data Engineer")]
```

## Competencias

Core, complementarias e indicadores por nivel, y la vista acumulada desde L1.


```python
c = en.get_competencies("Backend Developer", "L3")
print(len(c.core), len(c.complementary), len(c.indicators))
print(c.indicators[0])

acc = en.get_accumulated("Backend Developer", 5)
print(len(acc.core), "competencias acumuladas de L1 a L5")
```

## Progresión de carrera


```python
path = en.get_career_path("Backend Developer", "L5")
print("dominados:", [e.level_number for e in path.mastered])
print("crecimiento:", [e.level_number for e in path.growth])

nxt = en.get_next_level("Backend Developer", "L3")
print(nxt.next.code, "→", nxt.new_competencies[0])

print(en.get_by_experience("Backend Developer", 5).level)
```

## Buscar y filtrar


```python
for r in en.search("data", limit=3):
    print(r.role, r.score, r.matched_in)

print([e.code for e in en.filter(category="Design", level=3)])
print([r.role for r in es.search("seguridad", limit=3)])
```

## Comparar


```python
cmp = en.compare_levels("Backend Developer", 3, 4)
print(len(cmp.added), "competencias nuevas de L3 a L4")

cmp = en.compare_roles("Data Engineer", "Analytics Engineer", 3)
print("solo Data Engineer:", cmp.only_a[:2])
print("solo Analytics Engineer:", cmp.only_b[:2])
```

## Exportar a JSON o Markdown


```python
print(en.export(en.get_competencies("Backend Developer", 3), "markdown")[:300])
```
